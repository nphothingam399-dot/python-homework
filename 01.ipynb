```python
class Person:
    def __init__(self, cid, name, lname):
        self.cid = cid
        self.name = name
        self.lname = lname

    def info(self):
        print("{}: {} {}".format(self.cid, self.name, self.lname))
```

```python
class Student(Person):
    def __init__(self, cid, name, lname):
        super().__init__(cid, name, lname)

    def hellteacher(self):
        print("you")
```

```python
person1 = Person("1234567891233", "Phisan", "Sookkhee")
person1.info()
```

```python
student = Student("1234567891233", "Phisan", "Sookkhee")
student.info()
student.hellteacher()
```
