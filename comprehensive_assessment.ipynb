1. ‡∏™‡∏£‡πâ‡∏≤‡∏á dictionary ‡∏Ç‡∏≠‡∏á‡∏ô‡∏±‡∏Å‡∏®‡∏∂‡∏Å‡∏©‡∏≤ 3 ‡∏Ñ‡∏ô ‡πÇ‡∏î‡∏¢‡πÅ‡∏ï‡πà‡∏•‡∏∞‡∏Ñ‡∏ô‡∏°‡∏µ ‡∏£‡∏´‡∏±‡∏™, ‡∏ä‡∏∑‡πà‡∏≠, ‡πÅ‡∏•‡∏∞ ‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô
2. ‡πÄ‡∏û‡∏¥‡πà‡∏°‡∏ô‡∏±‡∏Å‡∏®‡∏∂‡∏Å‡∏©‡∏≤‡∏Ñ‡∏ô‡∏ó‡∏µ‡πà 4 ‡∏•‡∏á‡πÉ‡∏ô dictionary
3. ‡∏≠‡∏±‡∏õ‡πÄ‡∏î‡∏ï‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô‡∏Ç‡∏≠‡∏á‡∏ô‡∏±‡∏Å‡∏®‡∏∂‡∏Å‡∏©‡∏≤‡∏Ñ‡∏ô‡πÉ‡∏î‡∏Ñ‡∏ô‡∏´‡∏ô‡∏∂‡πà‡∏á
4. ‡πÅ‡∏™‡∏î‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ô‡∏±‡∏Å‡∏®‡∏∂‡∏Å‡∏©‡∏≤‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î (‡πÉ‡∏´‡πâ‡πÉ‡∏ä‡πâ loop)
5. ‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô‡πÄ‡∏â‡∏•‡∏µ‡πà‡∏¢‡∏Ç‡∏≠‡∏á‡∏ô‡∏±‡∏Å‡∏®‡∏∂‡∏Å‡∏©‡∏≤‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î

**Expected Output:**
```

‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô‡πÄ‡∏â‡∏•‡∏µ‡πà‡∏¢: XX.XX

[‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡πÅ‡∏ï‡πà‡∏•‡∏∞‡∏Ñ‡∏ô]
```

````python
# ‡∏™‡∏£‡πâ‡∏≤‡∏á dictionary ‡∏ô‡∏±‡∏Å‡∏®‡∏∂‡∏Å‡∏©‡∏≤ 3 ‡∏Ñ‡∏ô
students = {
    "66001": {"‡∏ä‡∏∑‡πà‡∏≠": "‡∏™‡∏°‡∏ä‡∏≤‡∏¢", "‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô": 80},
    "66002": {"‡∏ä‡∏∑‡πà‡∏≠": "‡∏™‡∏°‡∏´‡∏ç‡∏¥‡∏á", "‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô": 90},
    "66003": {"‡∏ä‡∏∑‡πà‡∏≠": "‡∏™‡∏°‡∏®‡∏±‡∏Å‡∏î‡∏¥‡πå", "‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô": 70}
}


students["66004"] = {"‡∏ä‡∏∑‡πà‡∏≠": "‡∏™‡∏°‡πÉ‡∏à", "‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô": 85}

students["66002"]["‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô"] = 95

total_score = 0

print("‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ô‡∏±‡∏Å‡∏®‡∏∂‡∏Å‡∏©‡∏≤:")
for student_id, info in students.items():
    print("‡∏£‡∏´‡∏±‡∏™:", student_id, "‡∏ä‡∏∑‡πà‡∏≠:", info["‡∏ä‡∏∑‡πà‡∏≠"], "‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô:", info["‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô"])
    total_score += info["‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô"]

average = total_score / len(students)

print("\n‡∏ô‡∏±‡∏Å‡∏®‡∏∂‡∏Å‡∏©‡∏≤‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î:", len(students), "‡∏Ñ‡∏ô")
print("‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô‡πÄ‡∏â‡∏•‡∏µ‡πà‡∏¢:", format(average, ".2f"))
````

### ‡∏Ç‡πâ‡∏≠ 4: ‡∏ô‡∏¥‡∏û‡∏à‡∏ô‡πå‡∏ó‡∏≤‡∏á‡∏Ñ‡∏ì‡∏¥‡∏ï‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå (Mathematical Expressions)

‡πÉ‡∏´‡πâ‡πÄ‡∏Ç‡∏µ‡∏¢‡∏ô‡πÇ‡∏õ‡∏£‡πÅ‡∏Å‡∏£‡∏°‡πÅ‡∏õ‡∏•‡∏á‡∏ô‡∏¥‡∏û‡∏à‡∏ô‡πå‡∏ó‡∏≤‡∏á‡∏Ñ‡∏ì‡∏¥‡∏ï‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå‡∏ï‡πà‡∏≠‡πÑ‡∏õ‡∏ô‡∏µ‡πâ‡πÄ‡∏õ‡πá‡∏ô‡πÇ‡∏Ñ‡πâ‡∏î Python ‡πÅ‡∏•‡∏∞‡πÅ‡∏™‡∏î‡∏á‡∏ú‡∏•‡∏•‡∏±‡∏û‡∏ò‡πå:

1. **‡∏û‡∏∑‡πâ‡∏ô‡∏ó‡∏µ‡πà‡∏ß‡∏á‡∏Å‡∏•‡∏°:** A = œÄr¬≤ ‡πÇ‡∏î‡∏¢‡πÉ‡∏´‡πâ r = 5
2. **‡∏ô‡∏¥‡∏û‡∏à‡∏ô‡πå‡πÄ‡∏®‡∏©‡∏™‡πà‡∏ß‡∏ô:** (15 + 25) √ó 3 √∑ 2‚Å¥
3. **‡∏£‡∏≤‡∏Å‡∏ó‡∏µ‡πà 3:** ‚àõ64 (cube root of 64)
4. **‡∏£‡∏∞‡∏¢‡∏∞‡∏ó‡∏≤‡∏á:** d = ‚àö((x‚ÇÇ-x‚ÇÅ)¬≤ + (y‚ÇÇ-y‚ÇÅ)¬≤) ‡πÇ‡∏î‡∏¢‡πÉ‡∏´‡πâ (x‚ÇÅ,y‚ÇÅ) = (3,4) ‡πÅ‡∏•‡∏∞ (x‚ÇÇ,y‚ÇÇ) = (7,1)

**‡∏´‡∏°‡∏≤‡∏¢‡πÄ‡∏´‡∏ï‡∏∏:** ‡πÉ‡∏ä‡πâ `math` module ‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö pi ‡πÅ‡∏•‡∏∞ sqrt

**Expected Output:**
```
1. ‡∏û‡∏∑‡πâ‡∏ô‡∏ó‡∏µ‡πà‡∏ß‡∏á‡∏Å‡∏•‡∏° = 78.50
2. ‡∏ú‡∏•‡∏•‡∏±‡∏û‡∏ò‡πå‡∏ô‡∏¥‡∏û‡∏à‡∏ô‡πå = 7.50
3. ‡∏£‡∏≤‡∏Å‡∏ó‡∏µ‡πà 3 ‡∏Ç‡∏≠‡∏á 64 = 4.00
4. ‡∏£‡∏∞‡∏¢‡∏∞‡∏ó‡∏≤‡∏á = 5.00
```

````python
import math
r = 5
area = math.pi * r**2

expression = (15 + 25) * 3 / (2**4)

cube_root = 64 ** (1/3)

x1, y1 = 3, 4
x2, y2 = 7, 1
distance = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)

print("1. ‡∏û‡∏∑‡πâ‡∏ô‡∏ó‡∏µ‡πà‡∏ß‡∏á‡∏Å‡∏•‡∏° =", format(area, ".2f"))
print("2. ‡∏ú‡∏•‡∏•‡∏±‡∏û‡∏ò‡πå‡∏ô‡∏¥‡∏û‡∏à‡∏ô‡πå =", format(expression, ".2f"))
print("3. ‡∏£‡∏≤‡∏Å‡∏ó‡∏µ‡πà 3 ‡∏Ç‡∏≠‡∏á 64 =", format(cube_root, ".2f"))
print("4. ‡∏£‡∏∞‡∏¢‡∏∞‡∏ó‡∏≤‡∏á =", format(distance, ".2f"))
````

4. ‡∏Å‡∏≥‡∏´‡∏ô‡∏î‡πÄ‡∏Å‡∏£‡∏î ‡πÇ‡∏î‡∏¢‡πÉ‡∏ä‡πâ‡πÄ‡∏á‡∏∑‡πà‡∏≠‡∏ô‡πÑ‡∏Ç‡∏ï‡πà‡∏≠‡πÑ‡∏õ‡∏ô‡∏µ‡πâ:

**‡∏ï‡∏±‡∏ß‡∏≠‡∏¢‡πà‡∏≤‡∏á‡∏Å‡∏≤‡∏£‡∏ó‡∏≥‡∏á‡∏≤‡∏ô:**
```python
# ‡πÉ‡∏´‡πâ‡∏£‡∏±‡∏ö‡∏Ñ‡πà‡∏≤‡∏à‡∏≤‡∏Å‡∏ú‡∏π‡πâ‡πÉ‡∏ä‡πâ‡∏´‡∏£‡∏∑‡∏≠‡∏Å‡∏≥‡∏´‡∏ô‡∏î‡∏Ñ‡πà‡∏≤‡πÄ‡∏≠‡∏á
score1, score2, score3 = 75, 65, 80
```

**Expected Output (‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡∏ï‡∏±‡∏ß‡∏≠‡∏¢‡πà‡∏≤‡∏á‡∏î‡πâ‡∏≤‡∏ô‡∏ö‡∏ô):**

````python
score1, score2, score3 = 75, 65, 80

print("‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô‡∏ß‡∏¥‡∏ä‡∏≤‡∏ó‡∏µ‡πà 1:", score1)
print("‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô‡∏ß‡∏¥‡∏ä‡∏≤‡∏ó‡∏µ‡πà 2:", score2)
print("‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô‡∏ß‡∏¥‡∏ä‡∏≤‡∏ó‡∏µ‡πà 3:", score3)

average = (score1 + score2 + score3) / 3
print("‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô‡πÄ‡∏â‡∏•‡∏µ‡πà‡∏¢:", format(average, ".2f"))

if average >= 50 and score1 >= 40 and score2 >= 40 and score3 >= 40:
    result = "‡∏ú‡πà‡∏≤‡∏ô"
else:
    result = "‡πÑ‡∏°‡πà‡∏ú‡πà‡∏≤‡∏ô"

print("‡∏ú‡∏•‡∏Å‡∏≤‡∏£‡∏™‡∏≠‡∏ö:", result)

if average >= 80:
    grade = "A"
elif average >= 70:
    grade = "B"
elif average >= 60:
    grade = "C"
elif average >= 50:
    grade = "D"
else:
    grade = "F"

print("‡πÄ‡∏Å‡∏£‡∏î:", grade)
````

‡πÉ‡∏´‡πâ‡πÄ‡∏Ç‡∏µ‡∏¢‡∏ô‡πÇ‡∏õ‡∏£‡πÅ‡∏Å‡∏£‡∏°‡∏ó‡∏µ‡πà‡πÅ‡∏™‡∏î‡∏á‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç 1 ‡∏ñ‡∏∂‡∏á 50 ‡πÇ‡∏î‡∏¢‡πÉ‡∏ä‡πâ loop ‡πÅ‡∏•‡∏∞‡∏à‡∏±‡∏î‡∏Å‡∏≤‡∏£‡∏î‡∏±‡∏á‡∏ô‡∏µ‡πâ:
1. ‡∏ñ‡πâ‡∏≤‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç‡∏´‡∏≤‡∏£‡∏î‡πâ‡∏ß‡∏¢ 3 ‡∏•‡∏á‡∏ï‡∏±‡∏ß ‡πÉ‡∏´‡πâ‡∏Ç‡πâ‡∏≤‡∏°‡πÑ‡∏õ (‡πÉ‡∏ä‡πâ continue)
2. ‡∏ñ‡πâ‡∏≤‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç‡∏´‡∏≤‡∏£‡∏î‡πâ‡∏ß‡∏¢ 7 ‡∏•‡∏á‡∏ï‡∏±‡∏ß ‡πÉ‡∏´‡πâ‡πÅ‡∏™‡∏î‡∏á "LUCKY" ‡πÅ‡∏ó‡∏ô‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç

**‡∏ï‡∏±‡∏ß‡∏≠‡∏¢‡πà‡∏≤‡∏á‡πÄ‡∏°‡∏∑‡πà‡∏≠ n = 4:**

**‡∏´‡∏°‡∏≤‡∏¢‡πÄ‡∏´‡∏ï‡∏∏:** ‡πÉ‡∏ä‡πâ nested loop ‡πÉ‡∏ô‡∏Å‡∏≤‡∏£‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏£‡∏π‡∏õ‡πÅ‡∏ö‡∏ö‡∏ô‡∏µ‡πâ

### ‡∏Ç‡πâ‡∏≠ 8: ‡∏ü‡∏±‡∏á‡∏Å‡πå‡∏ä‡∏±‡∏ô‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì‡∏™‡∏ñ‡∏¥‡∏ï‡∏¥ (Function for Statistics)

‡πÉ‡∏´‡πâ‡πÄ‡∏Ç‡∏µ‡∏¢‡∏ô‡∏ü‡∏±‡∏á‡∏Å‡πå‡∏ä‡∏±‡∏ô `calculate_stats(numbers)` ‡∏ó‡∏µ‡πà:
1. ‡∏£‡∏±‡∏ö list ‡∏Ç‡∏≠‡∏á‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç‡πÄ‡∏õ‡πá‡∏ô parameter
2. ‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì‡πÅ‡∏•‡∏∞ return dictionary ‡∏ó‡∏µ‡πà‡∏°‡∏µ keys ‡∏ï‡πà‡∏≠‡πÑ‡∏õ‡∏ô‡∏µ‡πâ:
   - `min`: ‡∏Ñ‡πà‡∏≤‡∏ó‡∏µ‡πà‡∏ô‡πâ‡∏≠‡∏¢‡∏ó‡∏µ‡πà‡∏™‡∏∏‡∏î
   - `max`: ‡∏Ñ‡πà‡∏≤‡∏ó‡∏µ‡πà‡∏°‡∏≤‡∏Å‡∏ó‡∏µ‡πà‡∏™‡∏∏‡∏î
   - `sum`: ‡∏ú‡∏•‡∏£‡∏ß‡∏°‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î
   - `count`: ‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç
   - `average`: ‡∏Ñ‡πà‡∏≤‡πÄ‡∏â‡∏•‡∏µ‡πà‡∏¢ (rounded to 2 decimal places)

**‡∏ï‡∏±‡∏ß‡∏≠‡∏¢‡πà‡∏≤‡∏á‡∏Å‡∏≤‡∏£‡πÉ‡∏ä‡πâ:**
```python
scores = [78, 85, 92, 64, 55, 88, 71, 95, 42, 80]
result = calculate_stats(scores)
print(result)
```

**Expected Output:**
```
{'min': 42, 'max': 95, 'sum': 750, 'count': 10, 'average': 75.0}
```

````python
def calculate_stats(numbers):
    stats = {
        "min": min(numbers),
        "max": max(numbers),
        "sum": sum(numbers),
        "count": len(numbers),
        "average": round(sum(numbers) / len(numbers), 2)
    }
    return stats
scores = [78, 85, 92, 64, 55, 88, 71, 95, 42, 80]
result = calculate_stats(scores)
print(result)
````

### ‡∏Ç‡πâ‡∏≠ 9: ‡∏ü‡∏±‡∏á‡∏Å‡πå‡∏ä‡∏±‡∏ô‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö‡∏£‡∏´‡∏±‡∏™‡∏ú‡πà‡∏≤‡∏ô (Password Validation Function)

‡πÉ‡∏´‡πâ‡πÄ‡∏Ç‡∏µ‡∏¢‡∏ô‡∏ü‡∏±‡∏á‡∏Å‡πå‡∏ä‡∏±‡∏ô `check_password(password)` ‡∏ó‡∏µ‡πà‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö‡∏Ñ‡∏ß‡∏≤‡∏°‡∏õ‡∏•‡∏≠‡∏î‡∏†‡∏±‡∏¢‡∏Ç‡∏≠‡∏á‡∏£‡∏´‡∏±‡∏™‡∏ú‡πà‡∏≤‡∏ô:

**‡πÄ‡∏á‡∏∑‡πà‡∏≠‡∏ô‡πÑ‡∏Ç‡∏Å‡∏≤‡∏£‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö:**
1. ‡∏Ñ‡∏ß‡∏≤‡∏°‡∏¢‡∏≤‡∏ß‡∏ï‡πâ‡∏≠‡∏á‡∏≠‡∏¢‡πà‡∏≤‡∏á‡∏ô‡πâ‡∏≠‡∏¢ 8 ‡∏ï‡∏±‡∏ß‡∏≠‡∏±‡∏Å‡∏©‡∏£
2. ‡∏ï‡πâ‡∏≠‡∏á‡∏°‡∏µ‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç (0-9) ‡∏≠‡∏¢‡πà‡∏≤‡∏á‡∏ô‡πâ‡∏≠‡∏¢ 1 ‡∏ï‡∏±‡∏ß
3. ‡∏ï‡πâ‡∏≠‡∏á‡∏°‡∏µ‡∏ï‡∏±‡∏ß‡∏≠‡∏±‡∏Å‡∏©‡∏£‡∏û‡∏¥‡∏°‡∏û‡πå‡πÉ‡∏´‡∏ç‡πà (A-Z) ‡∏≠‡∏¢‡πà‡∏≤‡∏á‡∏ô‡πâ‡∏≠‡∏¢ 1 ‡∏ï‡∏±‡∏ß

**‡∏ü‡∏±‡∏á‡∏Å‡πå‡∏ä‡∏±‡∏ô‡∏à‡∏∞ return:**
- `True` ‡∏ñ‡πâ‡∏≤ password ‡∏õ‡∏•‡∏≠‡∏î‡∏†‡∏±‡∏¢
- `False` ‡∏ñ‡πâ‡∏≤ password ‡πÑ‡∏°‡πà‡∏õ‡∏•‡∏≠‡∏î‡∏†‡∏±‡∏¢ ‡∏û‡∏£‡πâ‡∏≠‡∏°‡∏Å‡∏±‡∏ö‡∏Ç‡πâ‡∏≠‡∏Ñ‡∏ß‡∏≤‡∏°‡∏ö‡∏≠‡∏Å‡πÄ‡∏´‡∏ï‡∏∏‡∏ú‡∏•

**‡∏ï‡∏±‡∏ß‡∏≠‡∏¢‡πà‡∏≤‡∏á‡∏Å‡∏≤‡∏£‡πÉ‡∏ä‡πâ:**
```python
check_password("Hello123")     # True, "‡∏£‡∏´‡∏±‡∏™‡∏ú‡πà‡∏≤‡∏ô‡∏õ‡∏•‡∏≠‡∏î‡∏†‡∏±‡∏¢"
check_password("hello")         # False
check_password("Hello12")       # False
check_password("HELLO123")      # True
```

````python
def check_password(password):
    if len(password) < 8:
        return False, "‡∏£‡∏´‡∏±‡∏™‡∏ú‡πà‡∏≤‡∏ô‡∏ï‡πâ‡∏≠‡∏á‡∏°‡∏µ‡∏≠‡∏¢‡πà‡∏≤‡∏á‡∏ô‡πâ‡∏≠‡∏¢ 8 ‡∏ï‡∏±‡∏ß‡∏≠‡∏±‡∏Å‡∏©‡∏£"
    has_digit = False
    for ch in password:
        if ch.isdigit():
            has_digit = True
            break
    if not has_digit:
        return False, "‡∏£‡∏´‡∏±‡∏™‡∏ú‡πà‡∏≤‡∏ô‡∏ï‡πâ‡∏≠‡∏á‡∏°‡∏µ‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç‡∏≠‡∏¢‡πà‡∏≤‡∏á‡∏ô‡πâ‡∏≠‡∏¢ 1 ‡∏ï‡∏±‡∏ß"
    has_upper = False
    for ch in password:
        if ch.isupper():
            has_upper = True
            break
    if not has_upper:
        return False, "‡∏£‡∏´‡∏±‡∏™‡∏ú‡πà‡∏≤‡∏ô‡∏ï‡πâ‡∏≠‡∏á‡∏°‡∏µ‡∏ï‡∏±‡∏ß‡∏≠‡∏±‡∏Å‡∏©‡∏£‡∏û‡∏¥‡∏°‡∏û‡πå‡πÉ‡∏´‡∏ç‡πà‡∏≠‡∏¢‡πà‡∏≤‡∏á‡∏ô‡πâ‡∏≠‡∏¢ 1 ‡∏ï‡∏±‡∏ß"

    return True, "‡∏£‡∏´‡∏±‡∏™‡∏ú‡πà‡∏≤‡∏ô‡∏õ‡∏•‡∏≠‡∏î‡∏†‡∏±‡∏¢"
print(check_password("Hello123"))
print(check_password("hello"))
print(check_password("Hello12"))
print(check_password("HELLO123"))
````

### ‡∏Ç‡πâ‡∏≠ 10: ‡πÇ‡∏õ‡∏£‡πÅ‡∏Å‡∏£‡∏°‡∏à‡∏±‡∏î‡∏Å‡∏≤‡∏£‡∏£‡πâ‡∏≤‡∏ô‡∏Ñ‡πâ‡∏≤ (Inventory Management System)

```
   ‡∏£‡∏≤‡∏Ñ‡∏≤: 30 ‡∏ö‡∏≤‡∏ó/‡∏ä‡∏¥‡πâ‡∏ô
   ‡∏Ñ‡∏á‡πÄ‡∏´‡∏•‡∏∑‡∏≠: 5 ‡∏ä‡∏¥‡πâ‡∏ô
   ‡∏°‡∏π‡∏•‡∏Ñ‡πà‡∏≤: 150 ‡∏ö‡∏≤‡∏ó

========================
‡∏°‡∏π‡∏•‡∏Ñ‡πà‡∏≤‡∏™‡∏¥‡∏ô‡∏Ñ‡πâ‡∏≤‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î: 625 ‡∏ö‡∏≤‡∏ó
```

````python
inventory = {}

def add_product(name, price, quantity):
    inventory[name] = {
        "price": price,
        "quantity": quantity
    }

def sell_product(name, quantity):
    if name not in inventory:
        print("‡πÑ‡∏°‡πà‡∏û‡∏ö‡∏™‡∏¥‡∏ô‡∏Ñ‡πâ‡∏≤‡∏ô‡∏µ‡πâ‡πÉ‡∏ô‡∏Ñ‡∏•‡∏±‡∏á")
        return

    if inventory[name]["quantity"] < quantity:
        print("‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏™‡∏¥‡∏ô‡∏Ñ‡πâ‡∏≤‡πÑ‡∏°‡πà‡πÄ‡∏û‡∏µ‡∏¢‡∏á‡∏û‡∏≠")
        return

    inventory[name]["quantity"] -= quantity

def show_inventory():
    print(" ‡∏£‡∏≤‡∏¢‡∏Å‡∏≤‡∏£‡∏™‡∏¥‡∏ô‡∏Ñ‡πâ‡∏≤")
    index = 1
    for name, info in inventory.items():
        value = info["price"] * info["quantity"]
        print(f"{index}. {name}")
        print(f"   ‡∏£‡∏≤‡∏Ñ‡∏≤: {info['price']} ‡∏ö‡∏≤‡∏ó/‡∏ä‡∏¥‡πâ‡∏ô")
        print(f"   ‡∏Ñ‡∏á‡πÄ‡∏´‡∏•‡∏∑‡∏≠: {info['quantity']} ‡∏ä‡∏¥‡πâ‡∏ô")
        print(f"   ‡∏°‡∏π‡∏•‡∏Ñ‡πà‡∏≤: {value} ‡∏ö‡∏≤‡∏ó\n")
        index += 1
    print("")

def total_value():
    total = 0
    for info in inventory.values():
        total += info["price"] * info["quantity"]
    return total


# ‡∏ï‡∏±‡∏ß‡∏≠‡∏¢‡πà‡∏≤‡∏á‡∏Å‡∏≤‡∏£‡πÉ‡∏ä‡πâ‡∏á‡∏≤‡∏ô
add_product("‡∏ô‡∏°", 25, 10)
add_product("‡∏Ç‡∏ô‡∏°‡∏õ‡∏±‡∏á", 15, 20)
add_product("‡πÑ‡∏Ç‡πà", 30, 5)

sell_product("‡∏ô‡∏°", 3)

show_inventory()
print(f"‡∏°‡∏π‡∏•‡∏Ñ‡πà‡∏≤‡∏™‡∏¥‡∏ô‡∏Ñ‡πâ‡∏≤‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î: {total_value()} ‡∏ö‡∏≤‡∏ó")
````
