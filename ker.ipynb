## ให้สร้าง class student สําหรับจำลองนักศึกษา
-รหัสนักศึกษา
-หมายเลขบัตร 13 หลัก
-ชื่อ
-นามสกุล
-อายุ
-เพศ



```python
class Student:
    def __init__(self, student_id, citizen_id, first_name, last_name, age, gender):
        self.student_id = student_id         
        self.citizen_id = citizen_id          
        self.first_name = first_name          
        self.last_name = last_name           
        self.age = age                       
        self.gender = gender                

    def show_info(self):
        print("ข้อมูลนักศึกษา")
        print(f"รหัสนักศึกษา: {self.student_id}")
        print(f"เลขบัตรประชาชน: {self.citizen_id}")
        print(f"ชื่อ-นามสกุล: {self.first_name} {self.last_name}")
        print(f"อายุ: {self.age} ปี")
        print(f"เพศ: {self.gender}")
```

```python
student1 = Student(
    student_id="6811111111",
    citizen_id="1234567891011",
    first_name="ณัฐวุฒิ",
    last_name="โพธิ์งาม",
    age=18,
    gender="ชาย"
)

student1.show_info()
```
